# This IPython notebook is a document for plotting correlation heat map
    by Xia and Butorovic for JGR-Atmospheres manuscript
    
Please note: 
    The sns heatmap plotting has a visual issue in matplotlib 3.1.1. Please run this code if matplotlib has been downgraded to 3.1.0. See https://github.com/mwaskom/seaborn/issues/1773 for more information. I run these codes on another computer, so I should import the r-value and p-value data generated from PySPLIT analysis.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
%matplotlib inline
plt.rcParams["font.family"] = "Arial"
plt.rcParams.update({"mathtext.default":"regular"})
```

Read r-value and p-value data for d18O analysis

```python
df_heat_ast=pd.read_csv("heatmap_new_ast.csv")
df_heat_ast[df_heat_ast>=0.1]=4 # mark non-significant as 4
df_heat_ast[df_heat_ast<0.01]=3 # mark p < 0.01 as 3
df_heat_ast[df_heat_ast<0.05]=2 # mark p < 0.05 as 2
df_heat_ast[df_heat_ast<0.1]=1 # mark p < 0.10 as 1
df_heat_ast=df_heat_ast.replace(4,"")
df_heat_ast=df_heat_ast.replace(3,"***")
df_heat_ast=df_heat_ast.replace(2,"**")
df_heat_ast=df_heat_ast.replace(1,"*")
labels=df_heat_ast.to_numpy()
```

The correlation heatmap for between d18O residual and dE/dN/dW/dS.

```python
import numpy as np; np.random.seed(0)
import seaborn as sns; sns.set()
df_heat=pd.read_csv("heatmap_new.csv")
plt.figure(figsize=(6,9))
ax = sns.heatmap(df_heat, vmin=-1, vmax=1 ,linewidths=0,
                 yticklabels=["All months","DJF","MAM","JJA","SON"]*4,
                 cbar_kws={"label": "Correlation coefficient","shrink":0.3,"pad":0.1},
                 annot = labels,
                 fmt = "",
                 annot_kws={"color":"black","size":15},cmap="RdBu_r")
for t in ax.texts:
    trans = t.get_transform()
    offs = matplotlib.transforms.ScaledTranslation(0, 0.13,
                    matplotlib.transforms.IdentityTransform())
    t.set_transform( offs + trans )


cbar = ax.collections[0].colorbar
cbar.ax.tick_params(length=4, width=1)
plt.tight_layout()

ax.xaxis.set_ticks_position("top")
ax.tick_params(axis="x",length=0)
plt.xticks(rotation=45,ha="left")
ax.vlines([2, 4, 6], *ax.get_ylim(),linewidths=0.5)
ax.hlines([5, 10, 15], *ax.get_xlim(),linewidths=0.5)
plt.text(0.8,20.6,"dE")
plt.text(2.8,20.6,"dN")
plt.text(4.8,20.6,"dW")
plt.text(6.8,20.6,"dS")
plt.text(8.3,3,"500 m",rotation=90)
plt.text(8.3,8.2,"1000 m",rotation=90)
plt.text(8.3,13.2,"1500 m",rotation=90)
plt.text(8.3,18.2,"2000 m",rotation=90)
plt.savefig("traj heat.pdf",bbox_inches="tight")
```
